## Identificando valores faltantes em dados tabulares
Vamos iniciar nosso estudo identificando dados faltantes em tabelas.

```python
import pandas as pd

# importa o csv a partir de uma url no github para files ate 25MB
df = pd.read_csv('https://raw.githubusercontent.com/lucas-lxx/ML2023-LucasAlmeidaBarbosa/master/2019/01-Pre-processamento_Limpeza_Analise_de_Dados/missing.csv')
df
```

```python
# Olha coluna a coluna e soma os lugares aos quais isnull() e 
# avaliado como verdadeiro
df.isnull().sum()
```

```python
# Cria um array numpy com todos os valores do DataFrame
df.values
```

## Eliminando amostras ou *features* com valores ausentes

```python
# Remove linhas que contém valores faltantes
df.dropna(axis=0)
```

```python
# Remove colunas que contém valores faltantes
df.dropna(axis=1)
```

```python
# Drop apenas nas linhas onde todas as colunas são NaN
df.dropna(how='all')
```

```python
# Dropa as linhas que possuem menos de 3 valores reais
df.dropna(thresh=3)
```

```python
# Dropa apenas linhas onde NaN aparece em uma coluna específica
df.dropna(subset=['C'])
```
